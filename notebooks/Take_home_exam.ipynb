```python
import numpy as np
import pandas as pd
import matplotlib
import altair as alt
```

```python
covid = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv", parse_dates=['date'])
covid.shape
```

```python
covid = covid.rename(columns = {'location' :'Country'}, inplace = False) #renaming the column 'location' to 'Country'

#creating a dataframe with only the columns I require
new_covid = covid[['Country','date','continent', 'population', 'total_vaccinations_per_hundred',]]#Grouping all values by the 'Country' column, choosing the first value from its multiple rows and then reseting the index
new_covid= new_covid.groupby('Country').first().reset_index()#Removing the rows that have na/none in 'continent' column
new_covid=new_covid[new_covid['continent'].notna()].reset_index()#dropping the index column
new_covid= new_covid.drop('index', axis=1)

#dropping the date column
new_covid= new_covid.drop('date', axis=1)#removing spaces at the beginning and at the end
new_covid['Country'] = new_covid['Country'].str.strip()
```

```python
alt.Chart(new_covid).mark_bar().encode(
    y='total_vaccinations_per_hundred',
    x='Country'
)
```
